```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
index=range(0, 10001, 1000) #u_star
len_index = len(index)
columns=range(100, 1001, 100) #sigma_u
len_columns = len(columns)
mu, nu, sigma, u, sigma_u = 33.03, 45, 0.05, 5000, 500
```

```python
df = pd.DataFrame(np.arange(0,len_index*len_columns).reshape(len_index, len_columns), 
                  index=index, 
                  columns=columns)
df
```

```python
#I am a lazy student so I do not wanna use for loop
def find_index(element): #u_star
    return index[element // 10]
def find_axis(element):  #sigma_u
    return columns[element % 10]      
```

```python
calculator = lambda x: (nu + mu)/2 + (u - find_index(x)/2) * sigma / (2 * find_axis(x))
```

```python
#Question 3
#Index name related to u_star
#Columns name related to sigma_u
df.applymap(calculator)
```

```python
df.applymap(calculator).cov()
```
