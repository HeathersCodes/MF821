```python
#Question 1
'''
Download “minute bar” data (in the folder: eq_trades_1min) into Python pandas data frames for 
AAPL, BAC, and SPY for the two-month period from Feb 19th-April 17th, 2020 
( include both begin and end dates). One dataframe for each symbol.
'''
def data_processing(start_date: str, end_date: str, initial: str, name: str):
    start_date = datetime.strptime(start_date, '%Y%m%d').date()
    end_date = datetime.strptime(end_date, '%Y%m%d').date()
    current_date = start_date
    delta = timedelta(days=1)
    df_renew = pd.DataFrame()
    while current_date <= end_date:
        date_string = current_date.strftime('%Y%m%d')
        if os.path.isfile(os.path.join(path, f'{date_string}.zip')): #Check the file exist or not
            unzip_file = zipfile.ZipFile(path+f'{date_string}.zip', 'r') 
            open_file = unzip_file.open(f'{date_string}/{initial}/{name}.csv')
            df = pd.read_csv(open_file)
            df_renew = pd.concat([df_renew, df])
        current_date += delta
    return df_renew
```

```python
def sampleplot_and_qqplot(name, data):
    fig, ax = plt.subplots(nrows= 1, ncols=2, figsize=(10, 5))
    ax[0].set_xlim(-100, 100)
    ax[0].set_xlabel('1min returns (in bps)')
    ax[0].set_ylabel('Counts (in 1000s)')
    ax[0].hist(data*10000, bins=200)
    y_val = ax[0].get_yticks()
    ax[0].set_yticklabels([f'{x/1000: 3.0f}' for x in y_val])
    ax[0].set_title('Sample Distribution')

    # Plot the Q-Q plot
    #plt.figure(figsize=(8, 5))
    #ax = fig.add_subplot(1, 2, 1)
    stats.probplot(data, dist='norm', plot=plt)
    ax[1].set_title('QQ-plot')
    ax[1].set_ylabel('')
    #plt.show()
    return plt
```

```python
#Question 3
'''(3 pts) Please compute the r subscript min end subscript
(approximated by the 95% and 5% level values), alpha with hat on top 
for both the positive and negative tails of AAPL, BAC, and SPY for the 4 periods individually. 

Please report the result in a table in your Python notebook 
with symbols as column header, and the periods, right/left tails, and parameters as row header. 
'''
def estimate_power_law_parameters(returns, quantile=0.05):
    r_min, alpha_hat = [None,None], [None,None]
    r_min[0], r_min[1] = np.quantile(returns, quantile), np.quantile(returns, 1-quantile)
    pre_logret = [None,None]
    pre_logret[0] = returns[returns <= r_min[0]]/r_min[0]
    pre_logret[1] = returns[returns >= r_min[1]]/r_min[1]
    pre_logret[0] = pre_logret[0][pre_logret[0] != 0]
    pre_logret[1] = pre_logret[1][pre_logret[1] != 0]
    alpha_hat[0] = 1 + len(pre_logret[0]) / np.sum(np.log(pre_logret[0]))
    alpha_hat[1] = 1 + len(pre_logret[1]) / np.sum(np.log(pre_logret[1]))
    #alpha_hat = 1 + len(returns[returns >= r_min]) / np.sum(np.log(returns[returns >= r_min] / r_min))

    return [r_min, alpha_hat]
#test
estimate_power_law_parameters(returnSPY)
```

```python
results_df = pd.DataFrame({}, columns=['AAPL', 'BAC', 'SPY'], index=['Period',
                                                                     'Left Tail', 'Right Tail', 
                                                                     'alpha_hat_minus', 'alpha_hat_plus'])
results_df
```

```python
def returns(symbols, return_list):
    df = pd.DataFrame()
    for i in range(len(symbols)):
        return_list[i] = pd.Series(return_list[0], name=symbols[i])
        #return_list[i].rename({i: symbols[i]}, inplace=True)
        df = pd.concat([df, return_list[i]], axis=1)

    return df
```

```python
def create_table(period: str, symbols: list, returns, df=results_df):
    for i in range(len(symbols)):
        value = estimate_power_law_parameters(returns[i])
        df.iloc[:,i] = [0, value[0][0], value[0][1], value[1][0], value[1][1]]
    df.loc['Period'] = period
    return df
```
