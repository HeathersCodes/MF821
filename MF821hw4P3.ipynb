```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
index=range(0, 10001, 1000) #u_star
len_index = len(index)
columns=range(100, 1001, 100) #sigma_u
len_columns = len(columns)
mu, nu, sigma, u, sigma_u = 33.03, 45, 0.05, 5000, 500
```

```python
df = pd.DataFrame(np.arange(0,len_index*len_columns).reshape(len_index, len_columns), 
                  index=index, 
                  columns=columns)
df
```

```python
class Question3:
    def __init__(self, index=index, columns=columns, nu=nu, mu=mu, sigma=sigma, u=u, df=df):
        self.index = index
        self.columns = columns
        self.nu = nu
        self.mu = mu
        self.sigma = sigma
        self.u = u
        self.df = df

    def find_index(self, element):#u_star
        return self.index[element // 10]

    def find_axis(self, element):#sigma_u
        return self.columns[element % 10]
    
    #I am a lazy student so I do not wanna use for loop
    def calculator(self, x):
        index = self.find_index(x)
        axis = self.find_axis(x)
        return (self.nu + self.mu) / 2 + (self.u - index / 2) * self.sigma / (2 * axis)
```

```python
#Question 3
#Index name related to u_star
#Columns name related to sigma_u
table = Question3()
table.df = table.df.applymap(lambda x: table.calculator(x))
table.df
```
