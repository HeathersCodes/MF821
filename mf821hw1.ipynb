```python
from ibapi.client import *
from ibapi.wrapper import *
from ibapi.utils import iswrapper
import time
import numpy as np
import pandas as pd
```

```python
class TestApp(EClient, EWrapper):
    def __init__(self):
        EClient.__init__(self, self)
        self.df1 = pd.DataFrame(columns=['Date', 'high', 'Low', 'Close', 'Volume'])
        self.df2 = pd.DataFrame(columns=['Date', 'Ask', 'Bid'])
        self.df3 = pd.DataFrame(columns=['Date', 'high', 'Low', 'Close', 'Volume'])
        self.df4 = pd.DataFrame(columns=['Date', 'Ask', 'Bid'])
 
    def nextValidId(self, orderId: int):
         
        mycontract = Contract()
        mycontract.symbol = "AAPL"
        mycontract.secType = "STK"
        mycontract.exchange = "SMART"
        mycontract.currency = "USD"

        mycontract1 = Contract()
        mycontract1.symbol = "AMZN"
        mycontract1.secType = "STK"
        mycontract1.exchange = "SMART"
        mycontract1.currency = "USD"
 
        self.reqHistoricalData(1, mycontract, "20230202 11:59:00 US/Eastern", "1 D", "1 min", "TRADES", 1, 1, False, [])
        self.reqHistoricalData(2, mycontract, "20230202 11:59:00 US/Eastern", "1 D", "1 min", "BID_ASK", 1, 1, False, [])
        self.reqHistoricalData(3, mycontract1, "20230202 11:59:00 US/Eastern", "1 D", "1 min", "TRADES", 1, 1, False, [])
        self.reqHistoricalData(4, mycontract1, "20230202 11:59:00 US/Eastern", "1 D", "1 min", "BID_ASK", 1, 1, False, [])
        
    
    def historicalData(self, reqId, bar):
        if reqId == 1:
            self.df1.loc[len(self.df1)] = [bar.date, bar.high, bar.low, bar.close, bar.volume]
            
        elif reqId == 2:    
            self.df2.loc[len(self.df2)] = [bar.date, bar.high, bar.low]

        elif reqId == 3:
            self.df3.loc[len(self.df3)] = [bar.date, bar.high, bar.low, bar.close, bar.volume]
        else:
            self.df4.loc[len(self.df4)] = [bar.date, bar.high, bar.low]
            

    def historicalDataUpdate(self, reqId, bar):
        if reqId == 1:
            self.df1.loc[len(self.df1)] = [bar.date, bar.high, bar.low, bar.close, bar.volume]
        elif reqId == 2:
            self.df2.loc[len(self.df2)] = [bar.date, bar.high, bar.low]
        elif reqId == 3:
            self.df3.loc[len(self.df3)] = [bar.date, bar.high, bar.low, bar.close, bar.volume]
        else:
            self.df4.loc[len(self.df4)] = [bar.date, bar.high, bar.low]
            
    @iswrapper
    def historicalDataEnd(self, reqId, start, end):
        if reqId == 1: 
            return self.df1
        elif reqId == 2:
            return self.df2
        elif reqId == 3:
            return self.df3
        else:
            return self.df4
            
   

```

```python
app = TestApp()
app.connect("127.0.0.1", 7497, 100)
app.run()
#need to shut down the app after minutes
```

```python
#2
class TestApp(EClient, EWrapper):
    def __init__(self):
        EClient.__init__(self, self)
        self.df = pd.DataFrame(columns=['Trade Date', 'Commission Paid', 'Execution Price'])
 
    def nextValidId(self, orderId: OrderId):
         
        mycontract = Contract()
        mycontract.symbol = "AAPL"
        mycontract.secType = "STK"   
        mycontract.exchange = "SMART"
        mycontract.currency = "USD"
 
        self.reqContractDetails(orderId, mycontract)
         
    def contractDetails(self, reqId: int, contractDetails: ContractDetails):
        print(contractDetails.contract)
 
        myorder = Order()
        myorder.orderId = reqId
        myorder.action = "BUY"
        myorder.tif = "GTC"
        myorder.orderType = "MKT"
        myorder.totalQuantity = 100
 
        self.placeOrder(reqId, contractDetails.contract, myorder)
 
    @iswrapper
    def openOrder(self, orderId, contract, order, orderState):
        
        self.df['Commission Paid'] = [orderState.commission]
        #print(f"Commission Paid: {orderState.commission}")
        return self.df

    @iswrapper
    def execDetails(self, reqId, contract, execution):
        self.df['Trade Date'] = [execution.time]
        self.df['Execution Price'] = [execution.price]
        #print(f"Execution Price: {execution.price}")
        return self.df
 
 
app = TestApp()
app.connect("127.0.0.1", 7497, 100)
app.run()
```
